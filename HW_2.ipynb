```python
import numpy as np
```

```python
def equation(x):
    f = x**3 - 7 * x**2 + 14 * x - 5
    return f
e = 10 ** -8
xl = 0
xu = 1
count = 0
```

```python
def bisection(xl, xu, e):
    err = 1
    xr_old = 0
    count = 0
    while err > e:
        xr = (xl + xu)/2
        fl = equation(xl)
        fx = equation(xr)
        if count > 0:
            err = abs((xr - xr_old) / xr)
        if fx * fl < 0:
            xu = xr
        elif fx * fl > 0:
            xl = xr
        xr_old = xr    
        count = count +1
    return xr, err, count
print (bisection(0, 1, e))
```
