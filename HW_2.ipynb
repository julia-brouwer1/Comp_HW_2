```python
import numpy as np
from sympy import symbols, diff 
```

```python
def equation(x):
    f = x**3 - 7 * x**2 + 14 * x - 5
    return f
e = 10 ** -8
xl = 0
xu = 1
x0 = 0
count = 0
```

```python
def derivative(x):
    df = 3*x**2 - 14*x + 14
    return df
```

```python
def bisection(xl, xu, e):
    err = 1
    xr_old = 0
    count = 0
    while err > e:
        xr = (xl + xu)/2
        fl = equation(xl)
        fx = equation(xr)
        if count > 0:
            err = abs((xr - xr_old) / xr)
        if fx * fl < 0:
            xu = xr
        elif fx * fl > 0:
            xl = xr
        xr_old = xr    
        count = count +1
    return xr, count
print (bisection(0, 1, e))
```

```python
def NewtonRaphson(x0, e):
    x_old = x0
    count = 0
    while True:
        count += 1
        f_val = equation(x_old)
        df_val = derivative(x_old)
        x_new = x_old - f_val / df_val
        err = abs((x_new - x_old) / x_new)
        if err <= e:
            return x_new, count
        x_old = x_new
print (NewtonRaphson(x0, e))
```

```python
print (bisection(0, 1, e))
print (bisection(-1, 1, e))
print (bisection(0, 5, e))
print (bisection(-5, 5, e))
print (bisection(.3, .5, e))
```

```python
print (NewtonRaphson(x0, e))
print (NewtonRaphson(2, e))
print (NewtonRaphson(5, e))
print (NewtonRaphson(-1, e))
print (NewtonRaphson(-5, e))
```
