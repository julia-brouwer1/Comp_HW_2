```python
import numpy as np
import math
```

```python
def equation(x):
    f = x**3 - 7 * x**2 + 14 * x - 5
    return f
e = 10 ** -8
xl = 0
xu = 1
x0 = 0
count = 0
```

```python
def derivative(x):
    df = 3*x**2 - 14*x + 14
    return df
```

```python
def bisection(xl, xu, e):
    err = 1
    xr_old = 0
    count = 0
    while err > e:
        xr = (xl + xu)/2
        fl = equation(xl)
        fx = equation(xr)
        if count > 0:
            err = abs((xr - xr_old) / xr)
        if fx * fl < 0:
            xu = xr
        elif fx * fl > 0:
            xl = xr
        xr_old = xr    
        count = count +1
    return xr, count
print (bisection(0, 1, e))
```

```python
def NewtonRaphson(x0, e):
    x_old = x0
    count = 0
    while True:
        count += 1
        f_val = equation(x_old)
        df_val = derivative(x_old)
        x_new = x_old - f_val / df_val
        err = abs((x_new - x_old) / x_new)
        if err <= e:
            return x_new, count
        x_old = x_new
print (NewtonRaphson(x0, e))
```

```python
print (bisection(0, 1, e))
print (bisection(-1, 1, e))
print (bisection(0, 5, e))
print (bisection(-5, 5, e))
print (bisection(.3, .5, e))
```

```python
print (NewtonRaphson(x0, e))
print (NewtonRaphson(2, e))
print (NewtonRaphson(5, e))
print (NewtonRaphson(-1, e))
print (NewtonRaphson(-5, e))
```

```python
print (NewtonRaphson(1.7, e))
print (NewtonRaphson(2, e))
print (NewtonRaphson(1.00000000001, e))
```

```python
#Kepler's Equation

def Kepler(E, M, ecc):
    f = E - ecc * math.sin(E) - M
    return f
M0 = 1.0
ecc = 0.6
e = 10 ** -8
```

```python
def Kderivative(E, ecc):
    df = 1 - ecc * math.cos(E)
    return df
```

```python
def Kbisection(M, ecc, El, Eu, e):
    err = 1
    Er_old = 0
    count = 0
    while err > e:
        Er = (El + Eu)/2
        fl = Kepler(El, M, ecc)
        fx = Kepler(Er, M, ecc)
        if count > 0:
            err = abs((Er - Er_old) / Er)
        if fx * fl < 0:
            Eu = Er
        elif fx * fl > 0:
            El = Er
        Er_old = Er    
        count = count +1
    return Er, count
print (Kbisection(M0, ecc, 0, 5, e))
```

```python
def KNewtonRaphson(M, ecc, e):
    if ecc < 0.8:
        E = M
    else: 
        E = math.pi

    count = 0

    while True:
        count += 1
        f_val = Kepler(E, M0, ecc)
        df_val = Kderivative(E, ecc)
        
        E_new = E - f_val / df_val
        err = abs((E_new - E) / E_new)
        if err <= e:
            return E_new, count
        E = E_new
print (KNewtonRaphson(M0, ecc, e))
```
